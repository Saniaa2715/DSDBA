```python
# Scan all numeric variables for outliers

# Look for outliers in the age column

Q1 = data['Second year:   Sem 1'].quantile(0.25)
Q3 = data['Second year:   Sem 1'].quantile(0.75)

IQR = Q3 - Q1

print("Q1: ", Q1)
print("Q3: ", Q3)
print("IQR: ", IQR)

sns.boxplot(x= data['Second year:   Sem 1'])
```

```python
outliers = data[(data['Second year:   Sem 1'] < (Q1 - 1.5 * IQR)) | (data['Second year:   Sem 1'] > (Q3 + 1.5 * IQR))]
print(outliers)
```

```python
def removal_box_plot(df, column, threshold):
    sns.boxplot(df[column])
    plt.title(f'Original Box Plot of {column}')
    plt.show()

    removed_outliers = df[df[column] <= threshold]

    sns.boxplot(removed_outliers[column])
    plt.title(f'Box Plot without Outliers of {column}')
    plt.show()
    return removed_outliers


threshold_value = 8.5

no_outliers = removal_box_plot(data, 'Second year:   Sem 2', threshold_value)
print(no_outliers)
```
