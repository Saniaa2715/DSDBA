```python
#Importing necessary libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Loading the in-built dataset iris using seaborn

df=sns.load_dataset('iris')
```

```python
# 1)List down the features and their types (e.g., numeric, nominal) available in the dataset.

print("Features and data types:")

features = list(df.columns)
types = []
for feature in features:
  if df[feature].dtype == 'object':
    types.append('nominal')
  else:
    types.append('numeric')

for i, feature in enumerate(features):
  print(f"{i+1}. {feature}: {types[i]}")
```

```python
# 2) Create a histogram for each feature in the dataset to illustrate the feature distributions.

plt.title("Histogram of Sepal Length by Species")
sns.histplot(x="sepal_length",data=df,kde=True)
plt.show()

plt.title("Histogram of Sepal Width by Species")
sns.histplot(x="sepal_width",data=df,kde=True)
plt.show()

plt.title("Histogram of Petal Length by Species")
sns.histplot(x="petal_length",data=df,kde=True)
plt.show()

plt.title("Histogram of Petal Width by Species")
sns.histplot(x="petal_width",data=df,kde=True)
plt.show()


```

```python

# 3) Create a boxplot for each feature in the dataset
plt.title("Boxplot of Sepal Length by Species")
sns.boxplot(x="species",y="sepal_length",data=df)
plt.show()

plt.title("Boxplot of Sepal Width by Species")
sns.boxplot(x="species",y="sepal_width",data=df)
plt.show()

plt.title("Boxplot of Petal Length by Species")
sns.boxplot(x="species",y="petal_width",data=df)
plt.show()

plt.title("Boxplot of Petal Width by Species")
sns.boxplot(x="species",y="petal_length",data=df)
plt.show()





```

```python
# 4)Comparing distributions and identifying outliers

sns.boxplot(x='sepal_width', data=df)
plt.show()
sns.boxplot(x='sepal_length', data=df)
plt.show()
sns.boxplot(x='petal_width', data=df)
plt.show()
sns.boxplot(x='petal_length', data=df)
plt.figure(figsize=(5,5))
plt.show()
```
