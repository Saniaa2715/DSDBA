```python
# Defining X and Y

X = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = data['species']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
X
```

```python
# Splitting the data into training and testing model

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 42)

print("X train : ", X_train.shape)
print("X test : ", X_test.shape)
print("y train : ", y_train.shape)
print("y test : ", y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB

model=GaussianNB()
model.fit(X_train, y_train)
```

```python
print("The predictions on the test set are: ")
y_pred=model.predict(X_test)
y_pred
```

```python
res = pd.DataFrame({'Actual': y_test, 'Predicted' : y_pred})
res.head()
```

```python
# Computing Confusion Matrix

from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(y_test,y_pred)
conf_matrix
```

```python
sns.heatmap(pd.DataFrame(conf_matrix), annot=True, cmap = "hot",fmt='g')
plt.title("Confusion Matrix", y=1.1)
plt.ylabel("Actual")
plt.xlabel("Predicted")
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
class_labels = ["setosa", "virginica", "versicolor"]

# Flatten and extract TN, FP, FN, TP for each class
for i in range(len(class_labels)):
    TN = conf_matrix[i, i]
    FP = conf_matrix.sum(axis=0)[i] - conf_matrix[i, i]
    FN = conf_matrix.sum(axis=1)[i] - conf_matrix[i, i]
    TP = conf_matrix[i, i]

    print(f"\n{class_labels[i]}:")
    print("True Positives (TP):", TP)
    print("False Positives (FP):", FP)
    print("True Negatives (TN):", TN)
    print("False Negatives (FN):", FN)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Computing Accuracy

accuracy=accuracy_score(y_test, y_pred)
print("The Accuracy is : ", accuracy)
```

```python
# Computing Error Rate

error_rate= 1- accuracy
print("The Error rate is: ", error_rate)
```

```python
# Computing Precision

precision= precision_score(y_test, y_pred, average='weighted')
print("The Precision is: ", precision)
```

```python
# Computing Recall

recall= recall_score(y_test, y_pred,average='weighted')
print("The Recall is: ", recall)
```

```python
# Computing F1_score

F1_score=f1_score(y_test, y_pred,average='weighted')
print("The F1 score is: ", F1_score)
```
